---
### Load ratings dataframe

---

```python
rating_df = pd.read_csv('../data/clean/all_ratings.csv')
rating_df.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
rating_df.head()
```

---
### Get unique movies id

---

```python
rating_df_short = rating_df[['movieId','rating']]
score_by_movie = rating_df_short.groupby('movieId').mean()
print(score_by_movie.head())
```

```python
platform_df.set_index('id', inplace=True)
```

```python
platform_df = platform_df.assign(score = score_by_movie)
```

```python
# Promote id (index) to column again
platform_df = platform_df.reset_index()
platform_df
```

---
### Save the platform_df with the new score column to .csv

---

```python
import os

# Individual

root_path = '../data/clean/'
isExist = os.path.exists(root_path)
if not isExist:
    # Create a new directory because it does not exist
    os.makedirs(root_path)

platform_df.to_csv(root_path+"all_together_with_score.csv")
```
