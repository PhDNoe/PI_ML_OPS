### Load ratings dataframe

![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
rating_df = pd.read_csv('../data/clean/all_ratings.csv')
rating_df.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
rating_df.head()
```

![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)
### Get unique movies id

![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
rating_df_short = rating_df[['movieId','score']]
score_by_movie = rating_df_short.groupby('movieId').mean()


```

```python
platform_df.set_index('id', inplace=True)
```

```python
platform_df = platform_df.assign(score = score_by_movie)
```

```python
# Promote id (index) to column again
platform_df = platform_df.reset_index()
platform_df
```

![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)
### Duration_type has a typo

![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
platform_df['duration_type'].unique().tolist()
```

```python
my_typo_dict = {
    'min':'min',
    'season':'seasons',
    'seasons':'seasons'
}
platform_df['duration_type'] = platform_df['duration_type'].map(lambda x: my_typo_dict[x])
platform_df['duration_type'].unique().tolist()
```
